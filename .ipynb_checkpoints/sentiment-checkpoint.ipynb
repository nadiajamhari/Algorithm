```python
import plotly
import pandas as pd
from plotly.offline import  init_notebook_mode, plot, iplot
import plotly.graph_objects as go

stopwords = ['a', 'about', 'above', 'across', 'after', 'afterwards']
stopwords += ['again', 'against', 'all', 'almost', 'alone', 'along']
stopwords += ['already', 'also', 'although', 'always', 'am', 'among']
stopwords += ['amongst', 'amoungst', 'amount', 'an', 'and', 'another']
stopwords += ['any', 'anyhow', 'anyone', 'anything', 'anyway', 'anywhere']
stopwords += ['are', 'around', 'as', 'at', 'back', 'be', 'became']
stopwords += ['because', 'become', 'becomes', 'becoming', 'been']
stopwords += ['before', 'beforehand', 'behind', 'being', 'below']
stopwords += ['beside', 'besides', 'between', 'beyond', 'bill', 'both']
stopwords += ['bottom', 'but', 'by', 'call', 'can', 'cannot', 'cant']
stopwords += ['co', 'computer', 'con', 'could', 'couldnt', 'cry', 'de']
stopwords += ['describe', 'detail', 'did', 'do', 'done', 'down', 'due']
stopwords += ['during', 'each', 'eg', 'eight', 'either', 'eleven', 'else']
stopwords += ['elsewhere', 'empty', 'enough', 'etc', 'even', 'ever']
stopwords += ['every', 'everyone', 'everything', 'everywhere', 'except']
stopwords += ['few', 'fifteen', 'fifty', 'fill', 'find', 'fire', 'first']
stopwords += ['five', 'for', 'former', 'formerly', 'forty', 'found']
stopwords += ['four', 'from', 'front', 'full', 'further', 'get', 'give']
stopwords += ['go', 'had', 'has', 'hasnt', 'have', 'he', 'hence', 'her']
stopwords += ['here', 'hereafter', 'hereby', 'herein', 'hereupon', 'hers']
stopwords += ['herself', 'him', 'himself', 'his', 'how', 'however']
stopwords += ['hundred', 'i', 'ie', 'if', 'in', 'inc', 'indeed']
stopwords += ['interest', 'into', 'is', 'it', 'its', 'itself', 'keep']
stopwords += ['last', 'latter', 'latterly', 'least', 'less', 'ltd', 'made']
stopwords += ['many', 'may', 'me', 'meanwhile', 'might', 'mill', 'mine']
stopwords += ['more', 'moreover', 'most', 'mostly', 'move', 'much']
stopwords += ['must', 'my', 'myself', 'name', 'namely', 'neither', 'never']
stopwords += ['nevertheless', 'next', 'nine', 'no', 'nobody', 'none']
stopwords += ['noone', 'nor', 'not', 'nothing', 'now', 'nowhere', 'of']
stopwords += ['off', 'often', 'on','once', 'one', 'only', 'onto', 'or']
stopwords += ['other', 'others', 'otherwise', 'our', 'ours', 'ourselves']
stopwords += ['out', 'over', 'own', 'part', 'per', 'perhaps', 'please']
stopwords += ['put', 'rather', 're', 's', 'same', 'see', 'seem', 'seemed']
stopwords += ['seeming', 'seems', 'serious', 'several', 'she', 'should']
stopwords += ['show', 'side', 'since', 'sincere', 'six', 'sixty', 'so']
stopwords += ['some', 'somehow', 'someone', 'something', 'sometime']
stopwords += ['sometimes', 'somewhere', 'still', 'such', 'system', 'take']
stopwords += ['ten', 'than', 'that', 'the', 'their', 'them', 'themselves']
stopwords += ['then', 'thence', 'there', 'thereafter', 'thereby']
stopwords += ['therefore', 'therein', 'thereupon', 'these', 'they']
stopwords += ['thick', 'thin', 'third', 'this', 'those', 'though', 'three']
stopwords += ['three', 'through', 'throughout', 'thru', 'thus', 'to']
stopwords += ['together', 'too', 'top', 'toward', 'towards', 'twelve']
stopwords += ['twenty', 'two', 'un', 'under', 'until', 'up', 'upon']
stopwords += ['us', 'very', 'via', 'was', 'we', 'well', 'were', 'what']
stopwords += ['whatever', 'when', 'whence', 'whenever', 'where']
stopwords += ['whereafter', 'whereas', 'whereby', 'wherein', 'whereupon']
stopwords += ['wherever', 'whether', 'which', 'while', 'whither', 'who']
stopwords += ['whoever', 'whole', 'whom', 'whose', 'why', 'will', 'with']
stopwords += ['within', 'without', 'would', 'yet', 'you', 'your']
stopwords += ['yours', 'yourself', 'yourselves']


# Given a text string, remove all non-alphanumeric
# characters (using Unicode definition of alphanumeric).

def stripNonAlphaNum(file):
    import re
    words = (re.findall(r'\w+', (file).read().lower()))
    # new_items = [words for words in words if not words.isdigit()]
    return words

# Given a list of words, remove any that are
# in a list of stop words.

def listToString(s):
    # initialize an empty string
    str1 = ""

    # traverse in the string
    for ele in s:
        str1 += " "+ele+" "

    return str1

def removeStopwords(wordlist, stopwords):
    word = []
    test = listToString(stopwords)
    for w in wordlist:
        x = " "+w+" "
        result=BoyerMooreSearch(test,x)
        if(result != -1):
            word.append(w)

    return word
    # return [w for w in wordlist if w not in stopwords]


NO_OF_CHARS = 256

def badCharHeuristic(string, size):

    # Initialize all occurrence as -1
    badChar = [-1] * NO_OF_CHARS

    # set last position of each character in badChar
    for i in range(size):
        badChar[ord(string[i])] = i

    return badChar

def BoyerMooreSearch(list,w):
    m = len(w) #length of pattern
    n = len(list) #length of text

    badChar = badCharHeuristic(w, m)

    s=0  #shift
    while(s <= n-m):
        j = m - 1   #j = pattern length - 1, j is the last character first

        while j >= 0 and w[j] == list[s + j]: #check if j match
            j -= 1   #if match than reduce j, check left char pulak,  continuosly

        if j < 0: #maksudnya semua char match
            # s += (m - badChar[ord(list[s + m])] if s + m < n else 1)
            if (s + m < n): #if shift + pattern length kurang dari text length #teruskan while loop terus cari
                s += (m - badChar[ord(list[s + m])])  #
            else:
                s += 1
            return -1

        else: #tak match      #so shift based on bc
            s += max(1, j - badChar[ord(list[s + j])])



            # #to search from text
def binarySearch(list, w):
    start = 0
    end = len(list)-1
    while start <= end:
        middle = (start + end)//2
        midpoint = list[middle]
        if midpoint > w:
            end = middle -1
        elif midpoint < w:
            start = middle +1
        else:
            return -1 #exist




#to count no of words
def count(fullwordlist):
    total=0
    for w in fullwordlist:
        total=total+1
    return total

#to open file
def readFile(filename):
    name="Article/"+filename+".txt"
    file=open(name, encoding="utf8")
    return file

#to read file & ignore comma
def readNegativeWords():
    name="negative.txt"
    file=open(name, encoding="utf8")
    for line in file:
        field=line.split(",")
    file.close()
    # print(field)
    return field

#to read file & ignore comma
def readPositiveWords():
    name="positive.txt"
    file=open(name, encoding="utf8")
    for line in file:
        field=line.split(",")
    file.close()
    return field

#find and calculate -ve/+ve word from article text
def calculate(ve, wordlist):

    nTotal=0
    for w in wordlist:

        result = binarySearch(ve, w)
        if (result == -1):
            nTotal = nTotal + 1

    return nTotal

#calculate percentage -ve/+ve
def calculatePercentage(number,total):
    result=(number/total)*100
    return result

#compare percentage (which one is more)
def comparePercent(nPercent, pPercent):
    if nPercent>pPercent:
        return "negative"
    elif pPercent>nPercent:
        return "positive"
    else:
        return "equal"

#get percentage from total of -ve word + +ve word only
def comparePositiveNegative(nCount,pCount):
    conc=0
    if nCount>pCount:
        conc = (nCount/(pCount+nCount)) *100
    else:
        conc = (pCount/(pCount+nCount)) * 100
    return conc

def sortFreqDict(freqdict):
    aux = [(freqdict[key], key) for key in freqdict]
    aux.sort()
    aux.reverse()
    return aux

def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(list(zip(wordlist,wordfreq)))

def plotFreq(dictionary, file, indicator):
    state = ""
    if indicator == 0:
        state = "after"
    else:
        state = "before"
    x = []
    y = []

    for i, j in dictionary:
        x.append(j)
        y.append(i)

    fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers'))
    title = "Word Frequencies of " + file + " article " + state + " removing stop words"
    fig.update_layout(title=title,
                      xaxis_title='Word',
                      yaxis_title='Frequencies')

    fig.write_html("wordfreq/"+file + "_" + state + '.html')
```
